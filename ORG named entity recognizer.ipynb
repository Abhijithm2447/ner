```python
import requests
import json

import pandas as pd
from nltk.tokenize import sent_tokenize
from pprint import pprint
import pdb

# read data
FILE_NAME = "Institutions_for_NLP.csv"
data_csv = pd.read_csv(FILE_NAME, encoding = 'utf8', usecols=['Insight Text', 'Institutions'] )
mInsightText = data_csv['Insight Text']
mInstitutions = data_csv['Institutions']

def saveTrainingData(query, start,  end):
    BASE_URL = 'http://192.168.0.231:2424/'
    ENDPOINT = 'save_data2db/'

    context = {
      'query' :query,
      'entity_list' : [( start,  end)]
    }
    # new_emp = json.dumps(context)
    resp = requests.post(BASE_URL+ENDPOINT, data=context)
    print(resp.status_code)
    if resp.status_code == 200:
        res_json = resp.json()
        result = res_json['result']
        print("result: ", result)
        print("error: ")
        error = res_json['error']
        for er in error:
            print(er)

i = 0
# loop for each text
for text in mInsightText:
    i += 1
    # pdb.set_trace()
    # loop sentence
    for sent in sent_tokenize(text):
        
        org_l = mInstitutions[i-1]
        # loop for find multiple org
            
        for org in org_l.split(","):
            # search repeated org in sentence
            org = org.strip()
            if len(org) >0:
                    
                index_end = 0
                while True:
                    index = sent.find(org, index_end)
                    if(index == -1):
                        break
                    index_end = index + len(org)
                    # Train insight
    #                     print("org", org)
    #                     print("org filtered", sent[index:index_end])
                    print("-"*50)
                    saveTrainingData(sent, index, index_end)
                    print("-"*50)
                    print("sent:",sent)
                    print("org:",org)
                    print("-"*50)
```
